```python
with open("archivo.txt", "w") as f:
    f.write("Hola, este es un archivo de texto")

with open("archivo.txt", "r") as f:
    print("TXT:", f.read())
```

CSV File(s)

```python
import csv

with open("archivo.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Nombre", "Edad"])
    writer.writerow(["Diego", 18])

with open("archivo.csv", "r") as f:
    reader = csv.reader(f)
    print("CSV:")
    for fila in reader:
        print(fila)
```

JSON File(s)

```python
import json

datos = {
    "nombre": "Diego",
    "edad": 18
}

with open("archivo.json", "w") as f:
    json.dump(datos, f)

with open("archivo.json", "r") as f:
    print("JSON:", json.load(f))
```

CSV con Numpy

```python
import numpy as np
import pandas as pd
import json

print("NUMPY - CSV")
datos_numpy_csv = np.genfromtxt("archivo.csv", delimiter=",", dtype=str, encoding="utf-8")
print(datos_numpy_csv)

print("\n" + "="*40 + "\n")
```

CSV con Pandas

```python
print("PANDAS - CSV")
df_csv = pd.read_csv("archivo.csv")
print(df_csv)

print("\n" + "="*40 + "\n")
```

JSON con Numpy

```python
print("NUMPY - JSON")
with open("archivo.json", "r", encoding="utf-8") as f:
    datos_json = json.load(f)

datos_numpy_json = np.array(datos_json)
print(datos_numpy_json)

print("\n" + "="*40 + "\n")
```

JSON con Pandas

```python
print("PANDAS - JSON")

with open("archivo.json", "r", encoding="utf-8") as f:
    datos_json = json.load(f)

df_json = pd.DataFrame([datos_json])
print(df_json)
```

Datos Estructurados

CSV + Pandas

```python
import numpy as np
import pandas as pd
import json
from PIL import Image

print("CSV + PANDAS")
df_csv = pd.read_csv("archivo.csv")
print(df_csv)

print("\n" + "=" * 40 + "\n")
```

JSON + Pandas

```python
print("JSON + PANDAS")
with open("archivo.json", "r", encoding="utf-8") as f:
    datos_json = json.load(f)

df_json = pd.DataFrame([datos_json])
print(df_json)

print("\n" + "=" * 40 + "\n")
```

Datos No Estructurados

TXT + Numpy

```python
print("TXT + NUMPY")
with open("archivo.txt", "r", encoding="utf-8") as f:
    lineas = f.readlines()

txt_numpy = np.array(lineas)
print(txt_numpy)

print("\n" + "=" * 40 + "\n")
```
