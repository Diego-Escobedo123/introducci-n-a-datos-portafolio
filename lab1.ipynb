**Ejercicio 1**

* Leer archivos CSV y TXT usando context managers

```python
print("--- Leyendo archivo TXT con context manager ---")
try:
    with open('/content/sample_data/README.md', 'r') as file:
        content_txt = file.read()
        print("Contenido de 'README.md' (primeras 200 caracteres):\n")
        print(content_txt[:200])
except FileNotFoundError:
    print("Error: El archivo 'README.md' no se encontró.\n")
except Exception as e:
    print(f"Ocurrió un error al leer el archivo TXT: {e}\n")

print("--- Leyendo archivo CSV con context manager ---")
import csv
try:
    with open('/content/sample_data/california_housing_test.csv', 'r') as file:
        csv_reader = csv.reader(file)
        print("Primeras 6 filas de 'california_housing_test.csv' (incluyendo cabecera):\n")
        # Muestra la cabecera y las primeras 5 filas de datos
        for i, row in enumerate(csv_reader):
            print(row)
            if i >= 5: # La cabecera es i=0, las siguientes 5 filas son i=1 a i=5
                break
except FileNotFoundError:
    print("Error: El archivo 'california_housing_test.csv' no se encontró.\n")
except Exception as e:
    print(f"Ocurrió un error al leer el archivo CSV: {e}\n")
```

* Escribir archivos CSV y TXT usando context managers

```python
print("--- Escribiendo archivo TXT con context manager ---")
try:
    with open('salida.txt', 'w') as file:
        file.write('Esta es la primera línea.\n')
        file.write('Esta es la segunda línea del archivo TXT.\n')
    print("Archivo 'salida.txt' creado y escrito exitosamente.")
except Exception as e:
    print(f"Ocurrió un error al escribir el archivo TXT: {e}\n")

print("\n--- Escribiendo archivo CSV con context manager ---")
import csv

try:
    data = [
        ['Nombre', 'Edad', 'Ciudad'],
        ['Alice', 30, 'Nueva York'],
        ['Bob', 24, 'Los Ángeles'],
        ['Charlie', 35, 'Chicago']
    ]
    with open('salida.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(data)
    print("Archivo 'salida.csv' creado y escrito exitosamente.")
except Exception as e:
    print(f"Ocurrió un error al escribir el archivo CSV: {e}\n")
```

**¿En qué parte o partes de un Pipeline de Datos se pueden usar las operaciones de Lectura (R) y Escritura (W)?**

Las operaciones de lectura (R) y escritura (W) se utilizan a lo largo de todo un pipeline de datos: la lectura se aplica principalmente al inicio para obtener datos desde archivos, bases de datos o APIs, mientras que la escritura se usa al final para guardar los resultados procesados; sin embargo, ambas también pueden aparecer en la etapa intermedia de transformación, donde se leen datos temporales y se escriben versiones limpias o parciales conforme avanza el procesamiento.

**¿Se podría usar un DataFrame para todos los tipos de datos (Estructurados, No Estructurados, Semi Estructurados)?**

No, un DataFrame no se puede usar directamente para todos los tipos de datos. Funciona muy bien con datos estructurados como tablas y también puede adaptarse a datos semi estructurados si se organizan primero, pero no sirve directamente para datos no estructurados como imágenes o audio, ya que estos necesitan procesarse antes para poder representarse en forma de tabla.

**Ejercicio 3**

* Leer un archivo Pickle

```python
import pickle

pickle_file_name = 'mi_archivo.pkl'

try:
    # Crear un archivo pickle de ejemplo si no existe
    data_to_save = {'clave': 'valor', 'numero': 123, 'lista': [1, 2, 3]}
    with open(pickle_file_name, 'wb') as file:
        pickle.dump(data_to_save, file)
    print(f"Archivo '{pickle_file_name}' creado con datos de ejemplo.\n")

    # Ahora intentar cargar el archivo pickle
    with open(pickle_file_name, 'rb') as file:
        loaded_data = pickle.load(file)
    print(f"Datos cargados de '{pickle_file_name}' exitosamente:")
    print(loaded_data)
except Exception as e:
    print(f"Ocurrió un error al manipular el archivo Pickle: {e}")
```

* Escribir un archivo Pickle

```python
import pickle

output_pickle_file = 'mi_otro_archivo.pkl'
data_to_write = {
    'nombre': 'Carlos',
    'edad': 28,
    'ciudad': 'Madrid',
    'hobbies': ['leer', 'programar', 'caminar']
}

try:
    with open(output_pickle_file, 'wb') as file:
        pickle.dump(data_to_write, file)
    print(f"Datos escritos exitosamente a '{output_pickle_file}'.")
except Exception as e:
    print(f"Ocurrió un error al escribir el archivo Pickle: {e}")
```

**¿Qué tipo de dato se podría almacenar con un Pickle?**

Con Pickle se pueden almacenar prácticamente cualquier tipo de dato de Python, como diccionarios, listas, tuplas, números, strings e incluso objetos más complejos como clases o estructuras personalizadas.
